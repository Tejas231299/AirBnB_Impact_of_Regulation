```python
paris_listings_neighbourhood = paris_listings.groupby("neighbourhood"
        ).agg(avg_price = ("price","mean")).sort_values("avg_price",ascending = False).round(2)
```

```python
paris_listings_accommodates = paris_listings.groupby("accommodates")["price"].mean().sort_values(
    ascending = False).round(2)
```

```python
# Finding popularity of AirBnB over time
```

```python
paris_listings.head(10)
```

```python
paris_listings["year"] = paris_listings["host_since"].dt.year
```

```python
paris_listings.head()
```

```python
paris_listings_over_time = paris_listings.groupby("year").agg(total_new_hosts = ('year', 'count'),
    average_price = ('price', 'mean'))
```

```python
paris_listings_over_time
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_sorted = paris_listings_accommodates_Elysee.sort_values('avg_price',ascending=True)

x=df_sorted['avg_price'].values
y=df_sorted.index.astype(str)  

plt.figure(figsize=(12,8))
plt.barh(y=y,width=x,color='#00a699')

plt.title("Average Price of Airbnb in Elysee - vs - Number of people that can stay")
plt.xlabel("Average Price (Euros)")
plt.ylabel("Number of people that can stay")
plt.tight_layout()
plt.show()
```

```python
paris_listings_over_time
```

```python
plt.figure(figsize=(12,8))
sns.lineplot(data = paris_listings_over_time["total_new_hosts"],color="#fc642d",
            marker = "o")
plt.xlabel("years")
plt.ylabel("Number of Airbnbs and localities")
plt.title("popularity of airbnbs over time")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.lineplot(data = paris_listings_over_time["average_price"],color="#fc642d",
            marker = "o")
plt.xlabel("years")
plt.ylabel("Average Price")
plt.title("Average Price of AirBnBs  over time")
plt.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(paris_listings_over_time.index,paris_listings_over_time["total_new_hosts"],
        label = "New Hosts",c = "green")

ax.set_ylabel("New Hosts")
ax2 = ax.twinx()

ax2.plot(paris_listings_over_time.index,paris_listings_over_time["average_price"],
        label = "average_price",c = "red")
ax2.set_ylim(0)

ax2.set_ylabel("Average Price")

ax.set_title("Regulations Lead to Fewer New Hosts and Higher Prices")

plt.show()
```
